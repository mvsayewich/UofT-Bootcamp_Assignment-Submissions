```python
# Dependencies and Setup
import pandas as pd
import numpy as np
pd.options.display.float_format = '${0:,.2f}'.format

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
total_players = 0
average_price = 0
total_purchases = 0
total_revenue = 0
total_items = 0
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Get rid of all the empty values, if any
purchase_data.count()
# drop null rows
no_null_purchase_data = purchase_data.dropna(how='any')
no_null_purchase_data.head()
#Display the total number of players
total_players = no_null_purchase_data['SN'].nunique()
print(total_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Display the total number of unique items
total_items = no_null_purchase_data['Item Name'].nunique()
#print(total_items)
#Display the average price
average_price = round(no_null_purchase_data['Price'].mean(), 2)
#print(average_price)
#Calculate Number of Purchases
total_purchases = no_null_purchase_data['Purchase ID'].count()
#print(total_purchases)
#Calculate Total Revenue
total_revenue = no_null_purchase_data['Price'].sum()
#Create Summary Dataframe
summary_df = pd.DataFrame({"Total Players": [total_players],
                           "Number of Unique Items": [total_items],
                           "Number of Purchases": [total_purchases],
                           "Average Price": (average_price),
                           "Total Revenue": (total_revenue)})
summary_df.head()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create bins in which to place values based on ages
no_null_purchase_data = purchase_data.dropna(how='any')
purchase_analysis = no_null_purchase_data

bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"]

purchase_analysis["Age"] = pd.cut(purchase_analysis["Age"], bins, labels = group_labels) 
purchase_analysis.head()

total_people = purchase_analysis.groupby("Age")["SN"].nunique()
purchse_cnt = purchase_analysis.groupby("Age")["Price"].count()
avg_purchprice = purchase_analysis.groupby("Age")["Price"].mean()
total_purchvlu = purchase_analysis.groupby("Age")["Price"].sum()
avg_perperson = (total_purchvlu/total_people)
#print(total_people)
#print(purchse_cnt)
#print(avg_purchprice)
#print(total_purchvlu)
#print(avg_perperson)                                  


purchase_analysis_summary = pd.DataFrame({" ": ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"],
                                            "Purchase Count" : purchse_cnt,
                                            "Average Purchase Price": avg_purchprice, 
                                            "Total Purchase Value": total_purchvlu, 
                                            "Avg Total Purchase per Person Purchase" : avg_perperson})

pd.options.display.float_format = '${:,.2f}'.format 

#purchase_analysis_summary.sort_index(ascending=True) 

#Display Age Demographics Table
purchase_analysis_summary.set_index(" ", inplace = True)
purchase_analysis_summary
```
